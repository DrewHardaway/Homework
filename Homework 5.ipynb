````python
# Import NumPy, use one of the NumPy methods and create an array with a shape of (2, 3, 2). You can use the reshape method -- `.reshape()`
import numpy as np

arr = np.array(["A", "B", "C", "D", "a", "b", "c", "d", "e", "Andrew", "J", "Hardaway"])
rArr = arr.reshape(2,3,2)
print(rArr)

# 8. Use NumPy `.linspace()` to create an array with 6 linearly spaced values between 0 and 20
val = np.linspace(0, 20, 6)
print(val)

# 9. Make a Deep Copy of the above array
valDC = np.copy(val)
print(valDC)


# 10. Concatenate these 3 arrays into a new array named 'newArray'...
#     ```python
#             ([[25, 16]])
#             ([[11, 2], [13, 4]])
#             ([[7, 81], [5, 6], [11, 12]])
#     ```
arr2 = np.array([[25, 16]])
arr3 = np.array([[11, 2], [13, 4]])
arr4 = np.array([[7, 81], [5, 6], [11, 12]])

newArray = np.concatenate((arr2, arr3, arr4), axis=0)
print("newArray==", newArray)


# 11. Sort 'newArray' in order into 'sortedArray'
sortedArray = np.sort(newArray)

print(sortedArray)

# 12. Unpack the array tuples from the above 'reshapedArray'  into 4 well named variables. Print the 4 variables.

(a,b) = rArr

print("Unpacked array tuples=", a,b)

# 13. Combined and sort the following arrays into one called 'comboArray' ...

#     ```python
#         one = ([10, 11, 12, 13, 14, 15, 16, 17])
#         two = ([20, 21, 22, 23, 24, 25, 26, 27])
#         three = ([ 0, 1, 2, 3, 4, 5, 6, 7])
#     ```

array1 = np.array([10, 11, 12, 13, 14, 15, 16, 17])
array2 = np.array([20, 21, 22, 23, 24, 25, 26, 27])
array3 = np.array([ 0, 1, 2, 3, 4, 5, 6, 7])

comboA= np.concatenate((array1, array2, array3), axis=0)
comboArray= np.sort(comboA)
print("comboArray =", comboArray)

# 14. Take 'comboArray' and perform the following slicing activities:
#     - print sec1 - the 2nd element
#     - print sec2 - all elements from the 3rd element to the last
#     - print sec3 - all elements from the 4th to the 14th elements
#     - print sec4 - the last 6 elements
#     - print sec5 - all element from #0 up to and including #15, using the negative number method, i.e. taking a section from the end.
#     - print sec6 - from #20 every even element to the end
#     - print sec7 - from the last element moving forward, every 5th element.
sec1 = comboArray[1]
print(sec1)
sec2 = comboArray[2:]
print(sec2)
sec3 = comboArray[3:14]
print(sec3)
sec4= comboArray[:-6]
print(sec4)
sec5 = comboArray[:-10]
print(sec5)
sec6 = comboArray[16::2]
print(sec6)
sec7 = comboArray[::-5]
print(sec7)

````
